## 추천 알고리즘

```python
def recommend(location,near,food_cat,weight1,weight2,weight3,weight4):
  if (location is not "") & (near is "n") & (food_cat is not "") :
    sort1 = data[(data['핫플 위치'] == location) & (data['음식종류'] == food_cat)]
    if len(sort1) == 0 :
      return("해당 조건에 맞는 식당이 없습니다.")
    if weight1 == "" :
      print(sort1.nlargest(len(sort1), ['평점']))
    else:
      return (sort1.nlargest(len(sort1), [weight1, weight2, weight3, weight4],keep = "first"))

  if (location is not "") & (near is "y") & (food_cat is not "") :
    sector = data[data['핫플 위치'] == location]['카테고리'].iloc[0]
    sort2 = data[(data['카테고리'] == sector) & (data['음식종류'] == food_cat)]
    if len(sort2) == 0 :
      return("해당 조건에 맞는 식당이 없습니다.")
    if weight1 == "" :
      return(sort2.nlargest(len(sort2), ['평점']))
    else:
      return(sort2.nlargest(len(sort2), [weight1, weight2, weight3, weight4],keep = "first"))

  if (location is not "") & (near is "n") & (food_cat is "") :
    sort3 = data[(data['핫플 위치'] == location)]
    if len(sort3) == 0 :
      return("해당 조건에 맞는 식당이 없습니다.")
    if weight1 == "" :
      return(sort3.nlargest(len(sort3), ['평점']))
    else:
      return(sort3.nlargest(len(sort3), [weight1, weight2, weight3, weight4],keep = "first"))

  if (location is not "") & (near is "y") & (food_cat is "") :
    sector = data[data['핫플 위치'] == location]['카테고리'].iloc[0]
    sort4 = data[(data['카테고리'] == sector) & (data['음식종류'] == food_cat)]
    if len(sort4) == 0 :
      return("해당 조건에 맞는 식당이 없습니다.")
    if weight1 == "" :
      return(sort4.nlargest(len(sort4), ['평점']))
    else:
      return(sort4.nlargest(len(sort4), [weight1, weight2, weight3, weight4],keep = "first"))

  if (location is "") & (food_cat is not "") :
    sort5 = data[data['음식종류'] == food_cat]
    if len(sort5) == 0 :
      return("해당 조건에 맞는 식당이 없습니다.")
    if weight1 == "" :
      return(sort5.nlargest(len(sort5), ['평점']))
    else:
      return(sort5.nlargest(len(sort5), [weight1, weight2, weight3, weight4],keep = "first"))

  if (location is "") & (food_cat is "") :
    if weight1 == "" :
      return(data.nlargest(5, ['평점']))
    else:
      return(data.nlargest(5, [weight1, weight2, weight3, weight4],keep = "first"))
```

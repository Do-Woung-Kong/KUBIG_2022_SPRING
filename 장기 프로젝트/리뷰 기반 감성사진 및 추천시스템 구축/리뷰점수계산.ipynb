```python
review_dic = pd.DataFrame({'식당이름':df['title'],
                            '리뷰':df['review'],
                           '맛':taste,
                           '가격':price,
                           '서비스':service,
                           '분위기':mood })
review_dic
```

```python
null_list = []
```

```python
for j in range(len(review_dic)) :
  if( str(type(review_dic['리뷰'][j])) == "<class 'float'>" ) :
    null_list.append(j)
```

```python
review_dic = review_dic.drop(null_list)
review_dic
```

```python
review_dic.reset_index(drop=False,inplace=True)
review_dic
```

```python
for j in range(len(review_dic)) :
  for i in range(len(aa)):
    if (aa['expression'][i] in review_dic['리뷰'][j]) :

      if np.isnan(review_dic.loc[j,'맛']):
        review_dic.loc[j,'맛'] = aa['맛'][i]
      else : 
        review_dic.loc[j,'맛'] += aa['맛'][i]
      
      if np.isnan(review_dic.loc[j,'가격']):
        review_dic.loc[j,'가격'] = aa['가격'][i]
      else : 
        review_dic.loc[j,'가격'] += aa['가격'][i]

      if np.isnan(review_dic.loc[j,'서비스']):
        review_dic.loc[j,'서비스'] = aa['서비스'][i]
      else : 
        review_dic.loc[j,'서비스'] += aa['서비스'][i]
      
      if np.isnan(review_dic.loc[j,'분위기']):
        review_dic.loc[j,'분위기'] = aa['분위기'][i]
      else : 
        review_dic.loc[j,'분위기'] += aa['분위기'][i]

   

      
```

```python
def isEnglishOrKorean(input):
  k_count=0
  e_count=0
  for c in input:
    if ord('가')<= ord(c)<= ord('힣'):
      k_count+=1
    elif ord('a') <= ord(c.lower())<= ord('z'):
      e_count+=1
  return "한국어" if k_count>e_count else "영어"
```

```python
# 필요없는 단어 지우기
def word_delete(title):
    delete_word = [
        "1보","2보","3보","4보","5보","속보","사진",
        "게시판","주말 N","QA","그래픽","국내","동정",
        "신간","위클리","주간 화제의 뉴스","카드뉴스","팩트체크",
    ]
    for i in delete_word:
        if title.endswith(i) or title.startswith(i):
            title = title.replace(i, "")
    if title.endswith("종합"):
        title = title[:-2]
    title = title.replace("↑", " 증가")
    title = title.replace("↓", " 감소")
    title = title.replace("→", "에서 ")
    title = title.replace("~", "에서 ")
    title = title.replace("...", ". ")
    title = title.replace("·", " 그리고 ")
    title = title.replace(":", " 대 ")
    title = title.upper()
    return title
```

```python
# 불필요한 텍스트 제거
def clean_text(title):
    title_clean = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z0-9\\s.%]", " ", title)
    title_clean = re.sub(" +", " ", title_clean)
    title_clean = title_clean.strip()
    return title_clean
```

```python
def word_change(title):
    title = title.replace("%", "퍼센트 ")
    return title
```

```python
def word_eda(title):
    title = word_delete(title)
    title = chinese_char(title)
    title = clean_text(title)
    title = word_change(title)
    return title
```
